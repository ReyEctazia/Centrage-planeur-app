```python
import tkinter as tk
from tkinter import messagebox

def mm_to_percent(x_mm):
    return (x_mm - 220) / 140 * 100  # Plage 220-360 mm = 140 mm

def percent_to_mm(percent):
    return 220 + percent / 100 * 140

def calculer():
    try:
        # Données figées du Ventus 2a F-CJPR
        M_planeur = 240.4
        x_planeur_mm = 632
        M_parachute = 7
        x_parachute_mm = 455
        M_bagages = 2
        x_bagages_mm = 455
        x_gueuses_mm = 1635

        # Plage optimale centrage (en mm)
        plage_min = 320
        plage_max = 340

        # Entrées utilisateur
        M_pilote = float(entry_M_pilote.get())
        x_pilote_mm = 455  # fixe
        ballast_ailes_l = float(entry_ballast_ailes.get())
        ballast_queue_kg = float(entry_ballast_queue.get())
        batterie_queue_kg = float(entry_batterie_queue.get())
        cg_cible_mm = float(entry_cg_cible.get())
        masse_gueuses = float(entry_masse_gueuses.get())

        # Conversion mm → m
        def mm_to_m(x): return x / 1000
        x_planeur = mm_to_m(x_planeur_mm)
        x_pilote = mm_to_m(x_pilote_mm)
        x_parachute = mm_to_m(x_parachute_mm)
        x_bagages = mm_to_m(x_bagages_mm)
        x_gueuses = mm_to_m(x_gueuses_mm)
        x_ballast_ailes = mm_to_m(179)
        x_ballast_queue = mm_to_m(3900)
        x_batterie_queue = mm_to_m(3900)
        cg_cible = mm_to_m(cg_cible_mm)

        # Calcul moments
        moment_arriere = (
            M_planeur * x_planeur +
            ballast_ailes_l * x_ballast_ailes +
            ballast_queue_kg * x_ballast_queue +
            batterie_queue_kg * x_batterie_queue
        )
        moment_avant = (
            M_pilote * x_pilote +
            M_parachute * x_parachute +
            M_bagages * x_bagages +
            masse_gueuses * x_gueuses
        )

        # Masse totale et moment total
        M_tot = (
            M_planeur + M_pilote + M_parachute + M_bagages +
            ballast_ailes_l + ballast_queue_kg + batterie_queue_kg + masse_gueuses
        )
        moment_total = moment_arriere - moment_avant

        # Centrage actuel
        cg_actuel_m = moment_total / M_tot
        cg_actuel_mm = cg_actuel_m * 1000
        cg_actuel_percent = mm_to_percent(cg_actuel_mm)

        # Cible en %
        cg_cible_percent = mm_to_percent(cg_cible_mm)

        # Masse gueuses à ajouter ou retirer pour atteindre la cible
        masse_gueuses_necessaire = (moment_total - M_tot * cg_cible) / x_gueuses
        action = "Ajouter" if masse_gueuses_necessaire > 0 else "Retirer"

        # Texte résultat avec % mis en évidence
        texte_resultat = (
            f"à {plage_max} mm ({mm_to_percent(plage_max):.1f}%)\n\n"
            f"Centrage actuel : {cg_actuel_mm:.1f} mm ({cg_actuel_percent:.1f}%)\n"
            f"Centrage cible : {cg_cible_mm:.1f} mm ({cg_cible_percent:.1f}%)\n\n"
            f"{action} environ {abs(masse_gueuses_necessaire):.2f} kg de gueuses\n"
            f"pour atteindre le centrage cible souhaité.\n"
        )

        if plage_min <= cg_actuel_mm <= plage_max:
            texte_resultat += "\n✅ Le centrage est dans la plage optimale."
        else:
            texte_resultat += "\n⚠️ Le centrage est hors de la plage optimale."

        label_resultat.config(text=texte_resultat)

    except ValueError:
        messagebox.showerror("Erreur", "Veuillez entrer des nombres valides dans tous les champs.")

# Calcul des % pour affichage constants
plage_min = 320
plage_max = 340

valeurs_figees = (
    "Données figées Ventus 2a F-CJPR :\n"
    "- Masse planeur : 240.4 kg | Bras planeur vide : 632 mm\n"
    "- Masse parachute : 7 kg | Bras parachute : 455 mm\n"
    "- Masse bagages : 2 kg | Bras bagages : 455 mm\n"
    "- Bras des gueuses : 1635 mm\n"
    f"Plage optimale de centrage = {plage_min} mm ({mm_to_percent(plage_min):.1f}%) - "
    f"{plage_max} mm ({mm_to_percent(plage_max):.1f}%)\n"
)

# Création de la fenêtre principale
root = tk.Tk()
root.title("Centrage Ventus 2a F-CJPR")

# Affichage des constantes figées avec % mis en évidence
tk.Label(root, text=valeurs_figees, justify="left", fg="blue").grid(row=0, column=0, columnspan=2, sticky="w", padx=5)

# Champs de saisie utilisateur
row = 1
tk.Label(root, text="Poids pilote (kg) :").grid(row=row, column=0, sticky="e")
entry_M_pilote = tk.Entry(root)
entry_M_pilote.grid(row=row, column=1); row += 1

tk.Label(root, text="Ballast ailes (litres) :").grid(row=row, column=0, sticky="e")
entry_ballast_ailes = tk.Entry(root)
entry_ballast_ailes.grid(row=row, column=1); row += 1

tk.Label(root, text="Ballast queue (kg) :").grid(row=row, column=0, sticky="e")
entry_ballast_queue = tk.Entry(root)
entry_ballast_queue.grid(row=row, column=1); row += 1

tk.Label(root, text="Batterie queue (kg) :").grid(row=row, column=0, sticky="e")
entry_batterie_queue = tk.Entry(root)
entry_batterie_queue.grid(row=row, column=1); row += 1

tk.Label(root, text="Masse gueuses déjà installée (kg) :").grid(row=row, column=0, sticky="e")
entry_masse_gueuses = tk.Entry(root)
entry_masse_gueuses.grid(row=row, column=1); row += 1

tk.Label(root, text="Centrage cible (mm) :").grid(row=row, column=0, sticky="e")
entry_cg_cible = tk.Entry(root)
entry_cg_cible.grid(row=row, column=1); row += 1

# Bouton calcul
btn_calculer = tk.Button(root, text="Calculer", command=calculer)
btn_calculer.grid(row=row, column=0, columnspan=2, pady=10); row += 1

# Zone résultat
label_resultat = tk.Label(root, text="", justify="left")
label_resultat.grid(row=row, column=0, columnspan=2)

# Lancement de la fenêtre
root.mainloop()
```
